```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LassoCV
```

```
data=pd.read_csv('02. Xdata.csv')
nominal = ['LOC', 'FUR', 'WATER', 'HK_SER']
ordinal = ['AMB', 'PROX_SCH', 'PROX_ROAD', 'PROX_MALL']

for feature in ordinal:
  value = data[feature]
  encoder = LabelEncoder()
  trans_f = encoder.fit_transform(value)
  data[feature] = trans_f
```

```
data
```

```
#pd.get_dummies(data['LOC'], prefix='Estate')
data = pd.concat([pd.get_dummies(data[nominal], prefix='Estate'), data], axis=1)
data.drop(nominal, axis=1, inplace=True)
# Disadvantage - dimensionality increases i.e features are increased
data
```

```
X = data.drop('PRICE', 1)
y = data["PRICE"]
regressor = LassoCV()
regressor.fit(X, y)
#in every iteration we find the weights(coefficients)
#print(regressor.coef_)
#all weights with zeros are dropped

#rewriting to a dataframe series
coef_col = pd.Series(regressor.coef_, index=X.columns)
#print(coef_col)

sorted_coef = coef_col.sort_values()

plt.rcParams['figure.figsize'] = (15.0, 11.0)
sorted_coef.plot(kind='barh')
```
